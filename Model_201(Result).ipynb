# ***STEP 2: DEFINE THE MAIN FUNCTION(ENGINE)***

```python
import cv2
import dlib
import mediapipe as mp
import numpy as np
import os
import bz2
import urllib.request
from skimage import feature
from scipy import ndimage
import math
from skimage.feature import graycomatrix, graycoprops

def describe_lbp(image, num_points, radius, eps=1e-7):
    lbp = feature.local_binary_pattern(image, num_points, radius, method="uniform")
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, num_points + 3), range=(0, num_points + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + eps)
    return hist, lbp

def extract_all_features(image_path):
    """
    Extracts 201 facial features from an image and returns them as a single vector.
    """
    try:
        # --- Load Image & Models ---
        if not os.path.exists(image_path): return []
        image = cv2.imread(image_path)
        if image is None: return []
        img_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        h, w, _ = image.shape

        mp_face_mesh = mp.solutions.face_mesh
        face_mesh = mp_face_mesh.FaceMesh(static_image_mode=True, max_num_faces=1, refine_landmarks=True)
        predictor_path = 'shape_predictor_68_face_landmarks.dat'
        if not os.path.exists(predictor_path):
            url = "http://dlib.net/files/shape_predictor_68_face_landmarks.dat.bz2"
            urllib.request.urlretrieve(url, predictor_path + ".bz2")
            with bz2.BZ2File(predictor_path + ".bz2", 'rb') as f_in:
                with open(predictor_path, 'wb') as f_out: f_out.write(f_in.read())
            os.remove(predictor_path + ".bz2")
        detector = dlib.get_frontal_face_detector()
        predictor = dlib.shape_predictor(predictor_path)
        face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        mouth_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_mcs_mouth.xml')
        left_ear_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_mcs_leftear.xml')
        right_ear_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_mcs_rightear.xml')

        features = {}

        # --- Run Detections ---
        results_mp = face_mesh.process(img_rgb)
        results_dlib = detector(img_gray)
        faces_haar = face_cascade.detectMultiScale(img_gray, scaleFactor=1.1, minNeighbors=5)

        # --- Feature Extraction ---

        # Feature 1 & 2: Delaunay Triangulation Area and Centroid Length
        total_area, total_length = 0.0, 0.0
        if results_mp.multi_face_landmarks:
            face_landmarks_mp = results_mp.multi_face_landmarks[0]
            mp_landmarks = [(int(lm.x * w), int(lm.y * h)) for lm in face_landmarks_mp.landmark]
            rect = (0, 0, w, h)
            subdiv = cv2.Subdiv2D(rect)
            if mp_landmarks:
                for p in mp_landmarks:
                    if 0 <= p[0] < w and 0 <= p[1] < h:
                        subdiv.insert(p)
                triangles = subdiv.getTriangleList()
                if triangles is not None:
                    landmarks_np_mp = np.array(mp_landmarks, dtype=np.int32).reshape((-1, 1, 2))
                    centroids = []
                    for t in triangles:
                        pt1_tuple, pt2_tuple, pt3_tuple = (int(t[0]), int(t[1])), (int(t[2]), int(t[3])), (int(t[4]), int(t[5]))
                        if cv2.pointPolygonTest(landmarks_np_mp, pt1_tuple, False) >= 0 and \
                           cv2.pointPolygonTest(landmarks_np_mp, pt2_tuple, False) >= 0 and \
                           cv2.pointPolygonTest(landmarks_np_mp, pt3_tuple, False) >= 0:
                            total_area += 0.5 * abs(pt1_tuple[0]*(pt2_tuple[1]-pt3_tuple[1]) + pt2_tuple[0]*(pt3_tuple[1]-pt1_tuple[1]) + pt3_tuple[0]*(pt1_tuple[1]-pt2_tuple[1]))
                            centroids.append(((t[0]+t[2]+t[4])//3, (t[1]+t[3]+t[5])//3))
                    for i in range(len(centroids) - 1):
                        total_length += np.linalg.norm(np.array(centroids[i]) - np.array(centroids[i+1]))
        features['total_area'] = np.round(total_area, 2)
        features['total_length'] = np.round(total_length, 2)

        # Features 3-34: 16x16 Grayscale Matrix Column Means and Stds
        img_resized = cv2.resize(img_gray, (16, 16)).astype(np.float32) / 255.0
        features['col_means'] = np.round(np.mean(img_resized, axis=0), 2).tolist()
        features['col_stds'] = np.round(np.std(img_resized, axis=0), 2).tolist()

        # Feature 35: Pupil Distance
        d_pupil = 0.0
        if results_dlib and len(results_dlib) > 0:
            landmarks_np_dlib = np.array([[p.x, p.y] for p in predictor(img_gray, results_dlib[0]).parts()])
            if landmarks_np_dlib.shape[0] >= 48:
                left_pupil_center_dlib = np.mean(landmarks_np_dlib[36:42], axis=0).astype(int)
                right_pupil_center_dlib = np.mean(landmarks_np_dlib[42:48], axis=0).astype(int)
                d_pupil = np.linalg.norm(left_pupil_center_dlib - right_pupil_center_dlib)
            elif results_mp and results_mp.multi_face_landmarks:
                 mp_landmarks = [(int(lm.x * w), int(lm.y * h)) for lm in results_mp.multi_face_landmarks[0].landmark]
                 if 474 < len(mp_landmarks) and 469 < len(mp_landmarks):
                      left_pupil_mp = np.array(mp_landmarks[474])
                      right_pupil_mp = np.array(mp_landmarks[469])
                      d_pupil = np.linalg.norm(left_pupil_mp - right_pupil_mp)
        features['d_pupil'] = np.round(d_pupil, 2)

        # Feature 36 & 37: Face Area & Perimeter (from Haar Cascade)
        face_area, face_perimeter, fw, fh = 0.0, 0.0, 0, 0
        if len(faces_haar) > 0:
             (fx, fy, fw, fh) = faces_haar[0]
             face_area, face_perimeter = float(fw * fh), float(2 * (fw + fh))
        features['face_area'] = face_area
        features['face_perimeter'] = face_perimeter

        # Feature 38: Triangle area (pupils & nose) (using dlib)
        tri_area = 0.0
        if results_dlib and len(results_dlib) > 0 and 'landmarks_np_dlib' in locals() and landmarks_np_dlib.shape[0] >= 68:
             left_pupil_center_dlib = np.mean(landmarks_np_dlib[36:42], axis=0)
             right_pupil_center_dlib = np.mean(landmarks_np_dlib[42:48], axis=0)
             nose_tip_dlib = landmarks_np_dlib[30]
             tri_area = 0.5 * abs(left_pupil_center_dlib[0]*(right_pupil_center_dlib[1]-nose_tip_dlib[1]) + right_pupil_center_dlib[0]*(nose_tip_dlib[1]-left_pupil_center_dlib[1]) + nose_tip_dlib[0]*(left_pupil_center_dlib[1]-right_pupil_center_dlib[1]))
        features['tri_area'] = np.round(tri_area, 2)

        # Features 39-42: Additional Distances (d1, d2, d3, d4)
        d1, d2, d3, d4 = 0.0, 0.0, 0.0, 0.0
        if fw > 0 and fh > 0:
             d4 = fh
             d1 = 0.9 * fw
             if not mouth_cascade.empty():
                  mouths = mouth_cascade.detectMultiScale(img_gray[fy+fh//2:fy+fh, fx:fx+fw], 1.1, 5)
                  d3 = mouths[0][2] if len(mouths) > 0 else 0.8 * fw
             else: d3 = 0.8 * fw
             if not left_ear_cascade.empty() and not right_ear_cascade.empty():
                 left_ears = left_ear_cascade.detectMultiScale(img_gray[fy:fy+fh, fx:fx+fw], 1.1, 5)
                 right_ears = right_ear_cascade.detectMultiScale(img_gray[fy:fy+fh, fx:fx+fw], 1.1, 5)
                 if len(left_ears)>0 and len(right_ears)>0:
                      left_ear_center = (fx + left_ears[0][0] + left_ears[0][2] // 2, fy + left_ears[0][1] + left_ears[0][3] // 2)
                      right_ear_center = (fx + right_ears[0][0] + right_ears[0][2] // 2, fy + right_ears[0][1] + right_ears[0][3] // 2)
                      d2 = np.linalg.norm(np.array(right_ear_center) - np.array(left_ear_center))
                 else: d2 = 1.1 * fw
             else: d2 = 1.1 * fw
        features['d1'], features['d2'], features['d3'], features['d4'] = np.round(d1,2), np.round(d2,2), np.round(d3,2), np.round(d4,2)

        # Feature 43: Eye-to-eye distance to face width ratio
        eye_to_face_width_ratio = 0.0
        if fw > 0 and d_pupil > 0: eye_to_face_width_ratio = d_pupil / fw
        features['eye_to_face_width_ratio'] = np.round(eye_to_face_width_ratio, 4)

        # Feature 44: Nose bridge to face height ratio (using dlib)
        nose_bridge_ratio = 0.0
        if results_dlib and len(results_dlib) > 0 and 'landmarks_np_dlib' in locals() and landmarks_np_dlib.shape[0] >= 31:
             face_dlib_height = results_dlib[0].height()
             nose_top_dlib = landmarks_np_dlib[27]
             nose_tip_dlib = landmarks_np_dlib[30]
             nose_bridge_length_dlib = np.linalg.norm(nose_top_dlib - nose_tip_dlib)
             if face_dlib_height > 0: nose_bridge_ratio = nose_bridge_length_dlib / face_dlib_height
        features['nose_bridge_ratio'] = np.round(nose_bridge_ratio, 4)

        # Feature 45: Interocular distance / nose width (using dlib)
        interocular_ratio = 0.0
        if results_dlib and len(results_dlib) > 0 and 'landmarks_np_dlib' in locals() and landmarks_np_dlib.shape[0] >= 43:
             left_eye_inner_dlib = landmarks_np_dlib[39]
             right_eye_inner_dlib = landmarks_np_dlib[42]
             interocular_dist_dlib = np.linalg.norm(left_eye_inner_dlib - right_eye_inner_dlib)
             nose_left_dlib = landmarks_np_dlib[31]
             nose_right_dlib = landmarks_np_dlib[35]
             nose_width_dlib = np.linalg.norm(nose_left_dlib - nose_right_dlib)
             if nose_width_dlib > 0: interocular_ratio = interocular_dist_dlib / nose_width_dlib
        features['interocular_ratio'] = np.round(interocular_ratio, 4)

        # Feature 46: Inner Canthus to Outer Canthus Distance (MediaPipe)
        inner_canthus_dist = 0.0
        if results_mp and results_mp.multi_face_landmarks:
            mp_landmarks = [(int(lm.x * w), int(lm.y * h)) for lm in results_mp.multi_face_landmarks[0].landmark]
            if all(idx < len(mp_landmarks) for idx in [33, 133]):
                inner_canthus_dist = math.dist(mp_landmarks[33], mp_landmarks[133])
        features['inner_canthus_dist'] = np.round(inner_canthus_dist, 2)

        # Feature 47: Eye Aspect Ratio (EAR) (MediaPipe)
        ear = 0.0
        if results_mp and results_mp.multi_face_landmarks:
            mp_landmarks = [(int(lm.x * w), int(lm.y * h)) for lm in results_mp.multi_face_landmarks[0].landmark]
            if all(idx < len(mp_landmarks) for idx in [33,160,144,133]):
                dist_h = np.linalg.norm(np.array(mp_landmarks[33]) - np.array(mp_landmarks[133]))
                if dist_h > 0: ear = np.linalg.norm(np.array(mp_landmarks[160]) - np.array(mp_landmarks[144])) / dist_h
        features['ear'] = np.round(ear, 2)

        # Feature 48: Distance from eye to eyebrow (MediaPipe)
        eye_eyebrow_dist = 0.0
        if results_mp and results_mp.multi_face_landmarks:
            mp_landmarks = [(int(lm.x * w), int(lm.y * h)) for lm in results_mp.multi_face_landmarks[0].landmark]
            if all(idx < len(mp_landmarks) for idx in [159, 70]):
                eye_eyebrow_dist = math.dist(mp_landmarks[159], mp_landmarks[70])
        features['eye_eyebrow_dist'] = np.round(eye_eyebrow_dist, 2)

        # Feature 49: Nose tip to chin distance / face height (dlib)
        nose_chin_face_height_ratio = 0.0
        if results_dlib and len(results_dlib) > 0 and 'landmarks_np_dlib' in locals() and landmarks_np_dlib.shape[0] >= 68:
             nose_tip_dlib = landmarks_np_dlib[30]
             chin_dlib = landmarks_np_dlib[8]
             face_dlib_height = results_dlib[0].height()
             nose_to_chin_dist = np.linalg.norm(nose_tip_dlib - chin_dlib)
             if face_dlib_height > 0: nose_chin_face_height_ratio = nose_to_chin_dist / face_dlib_height
        features['nose_chin_face_height_ratio'] = np.round(nose_chin_face_height_ratio, 4)

        # Feature 50: Mouth width to nose width ratio (MediaPipe)
        mouth_nose_ratio = 0.0
        if results_mp and results_mp.multi_face_landmarks:
            mp_landmarks = [(int(lm.x * w), int(lm.y * h)) for lm in results_mp.multi_face_landmarks[0].landmark]
            if all(idx < len(mp_landmarks) for idx in [61,291,102,331]):
                mouth_w_mp = math.dist(mp_landmarks[61], mp_landmarks[291])
                nose_w_mp = math.dist(mp_landmarks[102], mp_landmarks[331])
                if nose_w_mp > 0: mouth_nose_ratio = mouth_w_mp / nose_w_mp
        features['mouth_nose_ratio'] = np.round(mouth_nose_ratio, 2)

        # Feature 51: Mouth center to chin distance (MediaPipe)
        mouth_chin_dist = 0.0
        if results_mp and results_mp.multi_face_landmarks:
            mp_landmarks = [(int(lm.x * w), int(lm.y * h)) for lm in results_mp.multi_face_landmarks[0].landmark]
            if all(idx < len(mp_landmarks) for idx in [13,14,152]):
                mc_pt = ((mp_landmarks[13][0]+mp_landmarks[14][0])//2, (mp_landmarks[13][1]+mp_landmarks[14][1])//2)
                mouth_chin_dist = math.dist(mc_pt, mp_landmarks[152])
        features['mouth_chin_dist'] = np.round(mouth_chin_dist, 2)

        # Features 52-53: Facial Symmetry
        h_sym_score, e_sym_score = 0.0, 0.0
        if results_mp and results_mp.multi_face_landmarks:
            landmarks = [(lm.x * w, lm.y * h) for lm in results_mp.multi_face_landmarks[0].landmark]
            central_x = np.mean([landmarks[i][0] for i in [152, 4, 1] if i < len(landmarks)]) if [i for i in [152,4,1] if i < len(landmarks)] else w/2
            h_discrepancies, e_discrepancies = [], []
            symmetric_pairs = [(33, 263), (133, 362), (61, 291), (234, 454)] # Simplified pairs
            for l_idx, r_idx in symmetric_pairs:
                if all(i < len(landmarks) for i in [l_idx, r_idx]):
                    l_pt, r_pt = landmarks[l_idx], landmarks[r_idx]
                    h_discrepancies.append(abs(abs(l_pt[0] - central_x) - abs(r_pt[0] - central_x)))
                    mirrored_x = central_x + (central_x - l_pt[0])
                    e_discrepancies.append(math.dist(r_pt, (mirrored_x, l_pt[1])))
            face_w = faces_haar[0][2] if len(faces_haar) > 0 else w
            if face_w > 0:
                h_sym_score = 1 - (np.mean(h_discrepancies) / face_w) if h_discrepancies else 1.0
                e_sym_score = 1 - (np.mean(e_discrepancies) / face_w) if e_discrepancies else 1.0
        features['horizontal_symmetry_score'] = np.round(h_sym_score, 3)
        features['euclidean_symmetry_score'] = np.round(e_sym_score, 2)

        # Features 54-79: LBP of Forehead
        lbp_hist_forehead = [0.0] * 26
        if len(faces_haar) > 0:
            (fx,fy,fw,fh) = faces_haar[0]
            forehead_roi = img_gray[fy:fy+int(fh*0.3), fx:fx+fw]
            if forehead_roi.size > 0:
                lbp_hist_forehead, _ = describe_lbp(forehead_roi, 24, 8)
                lbp_hist_forehead = lbp_hist_forehead.tolist()
        features['lbp_hist_forehead'] = lbp_hist_forehead

        # Features 80-105: LBP of region between eyebrows
        lbp_hist_eyebrows = [0.0] * 26
        if results_dlib and len(results_dlib) > 0:
            landmarks_np_dlib = np.array([[p.x, p.y] for p in predictor(img_gray, results_dlib[0]).parts()])
            if landmarks_np_dlib.shape[0] >= 48:
                l_eb_outer, r_eb_outer = landmarks_np_dlib[17], landmarks_np_dlib[26] # Outer eyebrow points
                l_eb_inner, r_eb_inner = landmarks_np_dlib[21], landmarks_np_dlib[22] # Inner eyebrow points

                # Define ROI based on inner eyebrow points
                eb_roi_x1, eb_roi_x2 = min(l_eb_inner[0], r_eb_inner[0]), max(l_eb_inner[0], r_eb_inner[0])
                eb_roi_y1 = min(l_eb_inner[1], r_eb_inner[1]) - 10 # Extend slightly above
                eb_roi_y2 = max(l_eb_inner[1], r_eb_inner[1]) + 10 # Extend slightly below

                # Ensure ROI is within image bounds
                eb_roi_x1 = max(0, eb_roi_x1)
                eb_roi_y1 = max(0, eb_roi_y1)
                eb_roi_x2 = min(w, eb_roi_x2)
                eb_roi_y2 = min(h, eb_roi_y2)

                eyebrow_region = img_gray[eb_roi_y1:eb_roi_y2, eb_roi_x1:eb_roi_x2]

                if eyebrow_region.size > 0:
                    lbp_hist_eyebrows, _ = describe_lbp(eyebrow_region, 24, 8)
                    lbp_hist_eyebrows = lbp_hist_eyebrows.tolist()
        features['lbp_hist_eyebrows'] = lbp_hist_eyebrows

        # Features 106-137: Gabor Filters on Irises
        left_gabor, right_gabor = [0.0]*16, [0.0]*16
        if results_mp and results_mp.multi_face_landmarks:
            landmarks_iris = results_mp.multi_face_landmarks[0].landmark
            def get_gabor_features(iris_indices, img, w, h, kernels):
                feats = []
                if all(idx < len(landmarks_iris) for idx in iris_indices):
                    center = np.mean([(landmarks_iris[i].x*w, landmarks_iris[i].y*h) for i in iris_indices], axis=0)
                    radius = 15
                    y1, y2 = int(center[1]-radius), int(center[1]+radius)
                    x1, x2 = int(center[0]-radius), int(center[0]+radius)
                    y1,y2,x1,x2 = max(0,y1), min(h,y2), max(0,x1), min(w,x2) # Clamp bounds
                    iris_region = img[y1:y2, x1:x2]
                    if iris_region.size > 0:
                        for kern in kernels:
                            f_img = ndimage.convolve(iris_region.astype(float), kern)
                            feats.extend([np.mean(f_img), np.std(f_img), np.max(f_img), np.min(f_img)])
                return feats if len(feats) == 16 else [0.0]*16

            kernels = [cv2.getGaborKernel((21,21),4,t,10,0.5,0) for t in [0,np.pi/4,np.pi/2,3*np.pi/4]]
            left_gabor = get_gabor_features(range(474,478), img_gray, w, h, kernels)
            right_gabor = get_gabor_features(range(469,473), img_gray, w, h, kernels)

        features['left_gabor'] = left_gabor
        features['right_gabor'] = right_gabor

        # Features 138-161: GLCM of Temple
        glcm_feats_list = [0.0]*24
        if len(faces_haar) > 0:
            (fx,fy,fw,fh) = faces_haar[0]
            # Approximate temple ROI (right temple)
            temple_roi_x1 = fx + int(fw * 0.7)
            temple_roi_y1 = fy + int(fh * 0.1)
            temple_roi_w = int(fw * 0.2)
            temple_roi_h = int(fh * 0.3)

            temple_roi_x2 = temple_roi_x1 + temple_roi_w
            temple_roi_y2 = temple_roi_y1 + temple_roi_h

            # Ensure ROI is within image boundaries
            temple_roi_x1, temple_roi_y1, temple_roi_x2, temple_roi_y2 = max(0, temple_roi_x1), max(0, temple_roi_y1), min(w, temple_roi_x2), min(h, temple_roi_y2)

            temple_image = img_gray[temple_roi_y1:temple_roi_y2, temple_roi_x1:temple_roi_x2]

            if temple_image.size > 0:
                glcm = graycomatrix(temple_image.astype(np.uint8), [1], [0,np.pi/4,np.pi/2,3*np.pi/4], 256, symmetric=True, normed=True)
                props = ['contrast','dissimilarity','homogeneity','energy','correlation','ASM']
                glcm_feats_list = [item for prop in props for item in graycoprops(glcm, prop).ravel()]
        features['glcm'] = glcm_feats_list

        # Features 162-170: Facial Ratios (based on Haar and Dlib)
        features['forehead_face_ratio'], features['midface_face_ratio'], features['lower_face_ratio_haar'] = 0.0, 0.0, 0.0
        features['forehead_midface_ratio'], features['midface_lower_face_ratio'] = 0.0, 0.0
        features['mouth_face_width_ratio'] = 0.0
        features['nose_bridge_top_len_ratio'], features['nose_bridge_mid_len_ratio'], features['nose_bridge_mid_top_ratio'] = 0.0, 0.0, 0.0

        if len(faces_haar) > 0:
            (fx,fy,fw,fh) = faces_haar[0]
            if fh > 0:
                features['forehead_face_ratio'] = np.round((fh*0.3)/fh, 2)
                features['midface_face_ratio'] = np.round((fh*0.4)/fh, 2)
                features['lower_face_ratio_haar'] = np.round((fh*0.3)/fh, 2)
            if fh*0.4 > 0: features['forehead_midface_ratio'] = np.round((fh*0.3)/(fh*0.4), 2)
            if fh*0.3 > 0: features['midface_lower_face_ratio'] = np.round((fh*0.4)/(fh*0.3), 2)

        if results_dlib and len(results_dlib) > 0:
            face_dlib = results_dlib[0]
            landmarks_np_dlib = np.array([[p.x, p.y] for p in predictor(img_gray, face_dlib).parts()])
            if landmarks_np_dlib.shape[0] >= 68:
                face_w_dlib = face_dlib.width()
                mouth_w_dlib = np.linalg.norm(landmarks_np_dlib[48] - landmarks_np_dlib[54])
                if face_w_dlib > 0: features['mouth_face_width_ratio'] = np.round(mouth_w_dlib/face_w_dlib, 2)
                nose_top = landmarks_np_dlib[27]
                nose_tip = landmarks_np_dlib[30]
                nose_len = np.linalg.norm(nose_top - nose_tip)
                nose_bridge_top_left = landmarks_np_dlib[21]
                nose_bridge_top_right = landmarks_np_dlib[22]
                nose_bridge_top_w = np.linalg.norm(nose_bridge_top_left - nose_bridge_top_right)
                nose_bridge_mid_left = landmarks_np_dlib[32]
                nose_bridge_mid_right = landmarks_np_dlib[34]
                nose_bridge_mid_w = np.linalg.norm(nose_bridge_mid_left - nose_bridge_mid_right)

                if nose_len > 0: features['nose_bridge_top_len_ratio'] = np.round(nose_bridge_top_w/nose_len, 2)
                if nose_len > 0: features['nose_bridge_mid_len_ratio'] = np.round(nose_bridge_mid_w/nose_len, 2)
                if nose_bridge_top_w > 0: features['nose_bridge_mid_top_ratio'] = np.round(nose_bridge_mid_w/nose_bridge_top_w, 2)


        # Features 171-185: Jaw Curvature
        jaw_curvature_features = [0.0] * 15
        if results_dlib and len(results_dlib) > 0:
            landmarks_np_dlib = np.array([[p.x, p.y] for p in predictor(img_gray, results_dlib[0]).parts()])
            if landmarks_np_dlib.shape[0] >= 17:
                jawline_pts = landmarks_np_dlib[0:17]
                for i in range(1, len(jawline_pts) - 1):
                    p_prev, p_curr, p_next = jawline_pts[i-1], jawline_pts[i], jawline_pts[i+1]
                    v1, v2 = p_prev - p_curr, p_next - p_curr
                    if np.linalg.norm(v1) > 0 and np.linalg.norm(v2) > 0:
                        cosine_angle = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
                        jaw_curvature_features[i-1] = np.degrees(np.arccos(np.clip(cosine_angle, -1.0, 1.0)))
        features['jaw_curvature'] = [np.round(j, 2) for j in jaw_curvature_features]

        # Features 186-190: Mouth Symmetry
        mouth_symmetry_scores = [0.0] * 5
        if results_dlib and len(results_dlib) > 0 and 'landmarks_np_dlib' in locals() and landmarks_np_dlib.shape[0] >= 68:
            mouth_pts = landmarks_np_dlib[48:68]
            left_corner, right_corner = mouth_pts[0], mouth_pts[6]
            mouth_center_x, mouth_center_y = (left_corner[0] + right_corner[0]) / 2, (left_corner[1] + right_corner[1]) / 2
            mouth_symmetry_scores[0] = abs(left_corner[1] - right_corner[1])
            if mouth_pts.shape[0] >= 5: # Check if necessary points exist
                upper_lip_left_y, upper_lip_right_y = mouth_pts[2][1], mouth_pts[4][1]
                dist_upper_left = abs(upper_lip_left_y - mouth_center_y)
                dist_upper_right = abs(upper_lip_right_y - mouth_center_y)
                if dist_upper_right > 0: mouth_symmetry_scores[1] = dist_upper_left / dist_upper_right
            if mouth_pts.shape[0] >= 11: # Check if necessary points exist
                lower_lip_left_y, lower_lip_right_y = mouth_pts[10][1], mouth_pts[8][1]
                dist_lower_left = abs(lower_lip_left_y - mouth_center_y)
                dist_lower_right = abs(lower_lip_right_y - mouth_center_y)
                if dist_lower_right > 0: mouth_symmetry_scores[2] = dist_lower_left / dist_lower_right
            if mouth_pts.shape[0] >= 6: # Check if necessary points exist
                dist_horiz_upper_left = abs(mouth_pts[1][0] - mouth_center_x)
                dist_horiz_upper_right = abs(mouth_pts[5][0] - mouth_center_x)
                if dist_horiz_upper_right > 0: mouth_symmetry_scores[3] = dist_horiz_upper_left / dist_horiz_upper_right
            if mouth_pts.shape[0] >= 12: # Check if necessary points exist
                dist_horiz_lower_left = abs(mouth_pts[11][0] - mouth_center_x)
                dist_horiz_lower_right = abs(mouth_pts[7][0] - mouth_center_x)
                if dist_horiz_lower_right > 0: mouth_symmetry_scores[4] = dist_horiz_lower_left / dist_horiz_lower_right
        features['mouth_symmetry'] = [np.round(m, 2) for m in mouth_symmetry_scores]

        # Feature 191: Jaw Angle at Center
        jaw_center_angle = 0.0
        if results_dlib and len(results_dlib) > 0 and 'landmarks_np_dlib' in locals() and landmarks_np_dlib.shape[0] >= 10:
            p_prev, p_curr, p_next = landmarks_np_dlib[7], landmarks_np_dlib[8], landmarks_np_dlib[9]
            v1, v2 = p_prev - p_curr, p_next - p_curr
            if np.linalg.norm(v1) > 0 and np.linalg.norm(v2) > 0:
                cosine_angle = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
                jaw_center_angle = np.degrees(np.arccos(np.clip(cosine_angle, -1.0, 1.0)))
        features['jaw_center_angle'] = np.round(jaw_center_angle, 2)

        # Features 192-200: Jaw Symmetry
        jaw_symmetry_scores = [0.0] * 9
        if results_dlib and len(results_dlib) > 0 and 'landmarks_np_dlib' in locals() and landmarks_np_dlib.shape[0] >= 31:
            chin_point = landmarks_np_dlib[8]
            symmetric_jaw_pairs = [(7, 9), (6, 10), (5, 11), (4, 12), (3, 13), (2, 14), (1, 15), (0, 16)]
            distance_ratios = []
            horizontal_ratios = []
            if landmarks_np_dlib.shape[0] >= 31: # Check if nose tip landmark exists
                nose_tip_point = landmarks_np_dlib[30]
                midline_x = (chin_point[0] + nose_tip_point[0]) / 2
                for i, (left_idx, right_idx) in enumerate(symmetric_jaw_pairs):
                    if all(idx < landmarks_np_dlib.shape[0] for idx in [left_idx, right_idx]):
                        left_point, right_point = landmarks_np_dlib[left_idx], landmarks_np_dlib[right_idx]
                        dist_left, dist_right = np.linalg.norm(chin_point - left_point), np.linalg.norm(chin_point - right_point)
                        if dist_right > 0: distance_ratios.append(dist_left / dist_right)
                        dist_left_midline, dist_right_midline = abs(left_point[0] - midline_x), abs(right_point[0] - midline_x)
                        if dist_right_midline > 0: horizontal_ratios.append(dist_left_midline / dist_right_midline)
            jaw_symmetry_scores[:len(distance_ratios)] = [np.round(r, 3) for r in distance_ratios] # Round ratios
            if horizontal_ratios: jaw_symmetry_scores[8] = np.round(np.mean(horizontal_ratios), 3) # Round mean
        features['jaw_symmetry'] = jaw_symmetry_scores


        # Feature 201: Number of Golden Ratios
        golden_ratio_count = 0
        GOLDEN_RATIO = 1.618
        TOLERANCE = 0.1
        if results_dlib and len(results_dlib) > 0:
            face_dlib = results_dlib[0]
            face_height = face_dlib.height()
            face_width = face_dlib.width()
            if face_width > 0:
                r1 = face_height / face_width
                if abs(r1 - GOLDEN_RATIO) <= TOLERANCE:
                    golden_ratio_count += 1

            if 'landmarks_np_dlib' in locals() and landmarks_np_dlib.shape[0] >= 68:
                nose_top = landmarks_np_dlib[27]
                nose_tip = landmarks_np_dlib[30]
                nose_length = np.linalg.norm(nose_top - nose_tip)
                nose_left = landmarks_np_dlib[31]
                nose_right = landmarks_np_dlib[35]
                nose_width = np.linalg.norm(nose_left - nose_right)
                if nose_width > 0:
                    r2 = nose_length / nose_width
                    if abs(r2 - GOLDEN_RATIO) <= TOLERANCE:
                        golden_ratio_count += 1

                mouth_left = landmarks_np_dlib[48]
                mouth_right = landmarks_np_dlib[54]
                mouth_width = np.linalg.norm(mouth_left - mouth_right)
                if nose_width > 0:
                    r3 = mouth_width / nose_width
                    if abs(r3 - GOLDEN_RATIO) <= TOLERANCE:
                        golden_ratio_count += 1

                left_eye_pts_dlib = landmarks_np_dlib[36:42]
                right_eye_pts_dlib = landmarks_np_dlib[42:48]
                left_pupil = np.mean(left_eye_pts_dlib, axis=0)
                right_pupil = np.mean(right_eye_pts_dlib, axis=0)
                pupil_dist = np.linalg.norm(left_pupil - right_pupil)
                if mouth_width > 0:
                    r4 = pupil_dist / mouth_width
                    if abs(r4 - GOLDEN_RATIO) <= TOLERANCE:
                        golden_ratio_count += 1

        features['golden_ratio_count'] = golden_ratio_count


        # --- Assemble Final Vector ---
        final_feature_vector = [
            features.get('total_area', 0.0),
            features.get('total_length', 0.0),
            *features.get('col_means', [0.0]*16),
            *features.get('col_stds', [0.0]*16),
            features.get('d_pupil', 0.0),
            features.get('face_area', 0.0),
            features.get('face_perimeter', 0.0),
            features.get('tri_area', 0.0),
            features.get('d1', 0.0),
            features.get('d2', 0.0),
            features.get('d3', 0.0),
            features.get('d4', 0.0),
            features.get('eye_to_face_width_ratio', 0.0),
            features.get('nose_bridge_ratio', 0.0),
            features.get('interocular_ratio', 0.0),
            features.get('inner_canthus_dist', 0.0),
            features.get('ear', 0.0),
            features.get('eye_eyebrow_dist', 0.0),
            features.get('nose_chin_face_height_ratio', 0.0),
            features.get('mouth_nose_ratio', 0.0),
            features.get('mouth_chin_dist', 0.0),
            features.get('horizontal_symmetry_score', 0.0),
            features.get('euclidean_symmetry_score', 0.0),
            *features.get('lbp_hist_forehead', [0.0]*26),
            *features.get('lbp_hist_eyebrows', [0.0]*26),
            *features.get('left_gabor', [0.0]*16),
            *features.get('right_gabor', [0.0]*16),
            *features.get('glcm', [0.0]*24),
            features.get('forehead_face_ratio', 0.0),
            features.get('midface_face_ratio', 0.0),
            features.get('lower_face_ratio_haar', 0.0),
            features.get('forehead_midface_ratio', 0.0),
            features.get('midface_lower_face_ratio', 0.0),
            features.get('mouth_face_width_ratio', 0.0),
            features.get('nose_bridge_top_len_ratio', 0.0),
            features.get('nose_bridge_mid_len_ratio', 0.0),
            features.get('nose_bridge_mid_top_ratio', 0.0),
            *features.get('jaw_curvature', [0.0]*15),
            *features.get('mouth_symmetry', [0.0]*5),
            features.get('jaw_center_angle', 0.0),
            *features.get('jaw_symmetry', [0.0]*9),
            features.get('golden_ratio_count', 0)
        ]

        if len(final_feature_vector) > 201:
            final_feature_vector = final_feature_vector[:201]
        elif len(final_feature_vector) < 201:
            final_feature_vector.extend([0.0] * (201 - len(final_feature_vector)))

        return final_feature_vector

    except Exception as e:
        print(f"An error occurred during feature extraction: {e}")
        import traceback
        traceback.print_exc()
        return []
# run for 2 times
```

**1. Comprehensive Feature Extraction:** This function is responsible for calculating a wide variety of features that attempt to numerically describe a person's face. These features can be categorized as:

Geometric Features: Distances, areas, and ratios between different facial landmarks (e.g., pupil distance, eye aspect ratio, nose length).

Texture Features: Local Binary Patterns (LBP) and Gray-Level Co-occurrence Matrix (GLCM) are used to capture the texture of different facial regions like the forehead and temples.

Symmetry Features: Calculates scores to represent the horizontal and euclidean symmetry of the face.

Gabor Features: Used to analyze the texture of the irises.

Golden Ratio Features: Counts how many facial proportions adhere to the golden ratio.

**2. Multi-Library Integration:**  It demonstrates how to use multiple computer vision libraries together to achieve a complex task. It uses:
dlib for face detection and landmark prediction.

MediaPipe for face mesh detection.

OpenCV for image loading, manipulation, and Haar cascade-based detection.

scikit-image for LBP and GLCM calculations.

**3. Foundation for the First Matching Approach:** The 201-feature vector produced by this function is the basis for the first matching algorithm. This vector is what is later used with cosine similarity and LDA to find a match in the dataset.

## ***ABOUT LDA***

**What is LDA?**

At its core, LDA is a **dimensionality reduction** technique, much like its more famous cousin, Principal Component Analysis (PCA). However, while PCA's main goal is to find the directions of maximum variance in the data (without caring about class labels), LDA has a different objective:

**LDA's goal is to find the feature subspace that maximizes the separability between classes.**

In the context of this notebook, each "class" is a different person in the dataset. So, LDA tries to find a new set of features (a "subspace") that makes it as easy as possible to distinguish between, for example, "Person A" and "Person B".

**How does it work (in simple terms)?**

Imagine you have a scatter plot of data points for two different classes. LDA tries to find a new axis (a "linear discriminant") to project these data points onto, such that:

The distance between the means of the two classes is maximized. (It pushes the class clusters as far apart as possible).

The variance within each class is minimized. (It makes each class cluster as "tight" and compact as possible).
By doing this, it creates a new, lower-dimensional representation of the data where the classes are well-separated.

**Why is it useful for facial recognition?**

**Improved Accuracy:**
 By focusing on the features that best separate individuals, LDA can often lead to more accurate matching results than using the original, high-dimensional features. It helps to filter out "noise" and focus on what makes each face unique.

**Reduced Computational Cost:**
 Working with a smaller number of features (as you can see from the output, it reduced the features from 201 to 47) makes the subsequent similarity calculations much faster.

**Overfitting Reduction:** By reducing the number of features, LDA can help to prevent the model from "memorizing" the training data and failing to generalize to new, unseen faces.

**Potential Downsides:**

**Linearity Assumption:**
As the name suggests, LDA is a linear method. It might not perform as well if the decision boundary between classes is highly non-linear.

**Sensitivity to Outliers:**
 LDA can be sensitive to outliers in the data.

**Need for Labeled Data:**
LDA is a supervised technique, meaning it requires the data to be labeled with the correct classes (in this case, the names of the individuals).

In summary, LDA is a powerful tool for improving the performance of classification and matching tasks by finding a more discriminative representation of the data. It's a great second step after initial feature extraction to refine the features and potentially boost the accuracy of the facial recognition system.

### **What are the 4096 Embeddings?**

The **4096 embeddings** are the heart of the deep learning approach to facial recognition. Think of the 4096-dimensional embedding as a highly detailed numerical "fingerprint" that uniquely represents a person's face. It's not a direct measurement of features like eye color or nose width. Instead, it's a dense vector of numbers that captures the subtle and complex characteristics of a face in a way that is highly discriminative.

**How are they created?**

The `VGG-Face` model is a deep convolutional neural network (CNN) that has been trained on millions of face images. As an image passes through the many layers of this network, the model learns to identify and encode increasingly complex features:

*   **Early Layers:** Might recognize simple things like edges, corners, and color gradients.
*   **Mid Layers:** Might combine these simple features to recognize more complex patterns like eyes, noses, and mouths.
*   **Deep Layers:** Combine these patterns to recognize entire faces and the subtle features that make each face unique.

The 4096-dimensional embedding is the output of one of the final layers of this network. It represents the model's high-level, abstract understanding of the face in the input image.

**Why are they so powerful?**

*   **Discriminative:** Faces that are similar will have embeddings that are close to each other in the 4096-dimensional space, while faces that are different will have embeddings that are far apart.
*   **Robust:** Because the model has been trained on so many images, the embeddings are often robust to changes in lighting, pose, expression, and even minor changes in appearance (like glasses or a new haircut). This is why they are so much more powerful than the manually engineered features.

**How are they used?**

Once we have the 4096-dimensional embedding for the new image and for all the images in the dataset, we can use a simple distance metric (like cosine similarity) to find the two embeddings that are closest to each other in the 4096-dimensional space. The face corresponding to the closest embedding is our best match.

    *   It uses **dlib** to detect the face and its 68 landmarks, which are essential for calculating ratios like nose bridge to face height and jaw curvature.
    *   It uses **OpenCV's Haar Cascades** for a more general detection of the face, mouth, and ears, which provides basic information like face area and perimeter.
    *   It uses **scikit-image** to calculate texture features like Local Binary Patterns (LBP) on the forehead and GLCM on the temples.
*   **Relevance:** This function is the "feature factory." It transforms a raw image into a structured numerical representation (the 201-feature vector) that can be used for mathematical comparison.

### **Step 3: Process the New Image**

*   **Purpose:** To take the new, unknown image (the "subject") and convert it into the same 201-feature vector format as the images in the database.
*   **Process:**
    1.  An image path is defined (`/content/8c485be7-146c-4d97-aa0d-de6eb6290f01.jpg`).
    2.  This path is passed to the `extract_all_features` function from Step 2.
    3.  The function returns the 201-feature vector for the new image, which is then printed to the screen.
*   **Relevance:** This step is crucial because you cannot compare an image to a database of feature vectors without first converting the image into a feature vector itself. This ensures that the comparison in Step 5 is "apples to apples."

### **Step 4: Load and Prepare the Database**

*   **Purpose:** To load the pre-computed facial features of over 550 known individuals and prepare them for comparison.
*   **Process:**
    1.  The `facial_features_final.csv` file, which contains the 201 features for each person, is loaded into a pandas DataFrame.
    2.  The names are separated from the numerical features.
    3.  **Crucially**, all feature values (both from the database and the new image) are scaled to a range between 0 and 1 using `MinMaxScaler`.
*   **Relevance:** Scaling is a critical data preprocessing step. It ensures that all 201 features contribute equally to the similarity calculation. Without scaling, a feature with a large range (like `face_area`) would dominate the calculation over a feature with a small range (like `eye_to_face_width_ratio`), leading to inaccurate results.

### **Step 5: The Matching Algorithm**

*   **Purpose:** To compare the feature vector of the new image against every feature vector in the database and find the most similar person.
*   **Process:**
    1.  It uses the **`cosine_similarity`** function from scikit-learn. This function calculates the cosine of the angle between the new image's 201-dimensional vector and each of the 550+ vectors in the database.
    2.  A similarity score close to 1.0 means the vectors are pointing in a very similar direction (high similarity), while a score close to 0.0 means they are nearly orthogonal (low similarity).
    3.  The scores are added to the DataFrame next to the corresponding names.
    4.  The DataFrame is sorted by the similarity score in descending order, and the top 5 matches are printed.
*   **Relevance:** This is the "moment of truth" for Stage 1. It's where the actual comparison happens and a ranked list of potential matches is generated based on the handcrafted features.

### **Step 6: Visualization**

# Stage 2: Enhanced Matching with Linear Discriminant Analysis (LDA)

This second stage of the model represents a significant refinement over the first. Instead of relying solely on the raw handcrafted features, it uses a powerful statistical technique called Linear Discriminant Analysis (LDA) to create a more discriminative and efficient feature set. The goal of this stage is to improve matching accuracy by focusing only on the features that best separate one individual from another.

### **Step 7.1: Data Preparation and Label Encoding**

*   **Purpose:** To prepare the scaled feature data from Stage 1 for use in the LDA model.
*   **Process:**
    1.  The scaled numerical features (the 201 features for each person) are converted into a NumPy array, which is the standard input format for scikit-learn models.
    2.  The names of the individuals are extracted. These will serve as the "labels" or "classes" for the LDA.
    3.  LDA is a supervised learning algorithm, meaning it needs to know which set of features belongs to which person. Since machine learning models work with numbers, not text, the `LabelEncoder` is used to convert each unique name into a unique integer (e.g., "POPESCU, ONEA MIRCEA FLORIN" might become 426, "BHAGADUR, RAMESH" might become 188, etc.).
*   **Relevance:** This is a critical preprocessing step. It formats the data correctly and provides the LDA algorithm with the necessary class labels so it can learn how to best distinguish between the different individuals in the dataset.

### **Step 7.2: Applying Linear Discriminant Analysis**

*   **Purpose:** To use LDA to transform the original 201 features into a new, smaller, and more powerful set of features.
*   **Process:**
    1.  An LDA model is initialized. `n_components` is the number of new features to create. By setting it to `None`, we let the algorithm determine the optimal number of components, which is typically the number of classes minus one. In this case, it reduced the feature space from 201 down to 47.
    2.  The `.fit_transform()` method does two things:
        *   **Fit:** It analyzes the relationship between the features and the class labels to find the "linear discriminants"—the new axes that maximize the separation between the classes while minimizing the variance within each class.
        *   **Transform:** It then projects the original 201-dimensional data onto these new 47 discriminant axes, creating a new dataset `X_lda` where each person is now represented by a 47-feature vector instead of a 201-feature vector.
    3.  The same transformation learned from the dataset is applied to the new image's feature vector. This is crucial to ensure the new image is represented in the same 47-dimensional space as the database images.
*   **Relevance:** This step is the essence of Stage 2. It intelligently reduces the complexity of the data, filtering out "noise" and retaining only the most discriminative information. This not only makes the subsequent matching process faster but also has the potential to make it more accurate by focusing on the features that truly matter for telling people apart.

### **Step 7.3: The Matching Algorithm with LDA Features**

*   **Purpose:** To perform the matching process again, but this time using the more refined 47-dimensional LDA features.
*   **Process:**
    1.  Just like in Stage 1, cosine similarity is used. However, this time it's calculating the similarity between the new image's 47-feature vector and the 47-feature vectors of everyone in the database.
    2.  The new scores are added to the DataFrame under the column `similarity_score_lda`.
    3.  The DataFrame is sorted by these new scores, and the top 5 matches are printed.
*   **Relevance:** This step demonstrates the impact of the LDA transformation. By comparing the results to those from Stage 1, you can see if focusing on the most discriminative features has changed or improved the ranking of the top matches.

### **Step 7.4: Visualization**

*   **Purpose:** To visually represent the results of the LDA-based matching.
*   **Process:**
    1.  It identifies the top 10 matches based on the new LDA similarity scores.
    2.  It creates a professional-looking bar chart that shows the names of the top 10 individuals on the y-axis and their corresponding LDA similarity scores on the x-axis.
    3.  The bar for the best match is highlighted, and the scores are annotated on the bars for clarity.
    4.  The input image is displayed alongside the chart for easy reference.
*   **Relevance:** This visualization provides a clear and immediate summary of the Stage 2 results. It allows for a quick comparison of the similarity between the top candidates and highlights the best match found by this more advanced statistical method.

# Stage 3: Advanced Matching with Deep Learning (VGG-Face Embeddings)

This final stage represents the state-of-the-art in facial recognition. It moves away from handcrafted features and statistical methods entirely, instead leveraging a powerful, pre-trained deep learning model called VGG-Face. The core idea is to use the model's "understanding" of faces, learned from millions of images, to generate a highly discriminative "embedding" (a numerical fingerprint) for each face.

### **Step 8.1: Setting Up the Deep Learning Environment**

*   **Purpose:** To prepare the environment for deep learning by installing the necessary library and loading the pre-trained VGG-Face model.
*   **Process:**
    1.  The code first unzips the folder containing the grayscale images of the individuals in the dataset. This is necessary because the deep learning model works directly with image files, not pre-extracted features from a CSV.
    2.  `!pip install deepface`: This installs the `deepface` library, a high-level Python package that wraps several state-of-the-art facial recognition models, making them easy to use.
    3.  `model = DeepFace.build_model('VGG-Face')`: This is the most critical part of the setup. It downloads the VGG-Face model architecture and its pre-trained weights. This model is a deep convolutional neural network (CNN) that has been trained on a massive dataset of faces and has learned to extract robust and discriminative features.
*   **Relevance:** This step provides the "brain" for Stage 3. By loading the pre-trained VGG-Face model, you are equipping the notebook with a powerful tool that can analyze and understand faces at a level far beyond what is possible with handcrafted features.

### **Step 8.2: Extracting Deep Features (Embeddings)**

*   **Purpose:** To use the VGG-Face model to generate a 4096-dimensional feature vector (an "embedding") for the new image and for every image in the dataset.
*   **Process:**
    1.  The code iterates through the list of names from the original CSV file. For each name, it finds the corresponding image file in the unzipped directory.
    2.  `DeepFace.represent()`: This function is the core of the feature extraction. It takes an image path, feeds the image through the VGG-Face model, and outputs a 4096-dimensional vector. This vector is the embedding—a rich, numerical representation of the face.
    3.  This process is repeated for the new input image, generating an embedding for it as well.
    4.  The code includes error handling for missing images or images with filenames that cause issues.
*   **Relevance:** This step replaces the entire manual feature extraction process of Stage 1. Instead of calculating 201 geometric and texture features, it uses the deep learning model to automatically generate a much more comprehensive and robust 4096-feature embedding. This embedding captures the subtle nuances of a face that are difficult or impossible to define manually.

### **Step 8.3: The Matching Algorithm with Deep Features**

*   **Purpose:** To compare the embedding of the new image with the embeddings of all the images in the dataset to find the best match.
*   **Process:**
    1.  The code first handles any potential errors from the previous step by replacing `NaN` (Not a Number) values with zeros.
    2.  `cosine_similarity()`: Just like in the previous stages, cosine similarity is used to find the similarity between the new image's 4096-dimensional embedding and the embeddings of all the database images.
    3.  The resulting scores are added to a DataFrame, and the best match (the one with the highest similarity score) is identified and printed.
*   **Relevance:** While the matching algorithm itself (cosine similarity) is the same, the data it operates on is fundamentally different and far more powerful. The comparison is now happening in a 4096-dimensional space that has been optimized by the deep learning model to separate faces effectively.

### **Step 8.4: Visualization**

*   **Purpose:** To present the results from the deep learning stage in a clear and visually appealing way.
*   **Process:**
    1.  Two types of visualizations are created:
        *   A horizontal bar chart showing the top 10 most similar individuals and their similarity scores, with the best match highlighted.
        *   A "lineup" visualization, similar to Stage 1, showing the input subject next to the top 4 matches found by the deep learning model, with the best match highlighted with a cyan border.
*   **Relevance:** These visualizations provide a final, compelling summary of the most advanced matching results. They make it easy to see who the deep learning model identified as the most probable match and how the other top candidates compare.
